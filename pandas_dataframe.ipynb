```python
#subplot
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.subplot(2,1,2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.title(" two lines in a plot")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.grid(True)
plt.show()
```

```python
student={'Bob':87,"Matt":56,"sam":27}
names=list(student.keys())
values=list(student.values())
plt.bar(names,values,color='r',linewidth=1)
plt.title("this is a bar plot")
plt.xlabel("this is name of student")
plt.ylabel("this is marks of student")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,values,color='g')
plt.title("barplot")
plt.xlabel("name")
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```
